```python
all_path = '/content/ALL_solidity_versions.txt' # all solidity releases

path1 = '/content/FINAL_smart_contract_data_collection.txt' #33767
path2 = '/content/FINAL_DApp.txt' #21243
```

```python
def read_txt(path):
  with open(path, 'r') as file:
      text = file.read()
  versions = text.split('\n')
  return versions
```

```python
all_vers = read_txt(all_path)
repo1_vers = read_txt(path1)
repo2_vers = read_txt(path2)
```

```python
print(len(all_vers))
print(len(repo1_vers))
print(len(repo2_vers))
print('0.6' in repo2_vers)
print(33767 + 21243)
```

```python
sol_version_dict = {}

# set up dict
for ver in all_vers:
  sol_version_dict[ver] = 0

### smart_contract_data
# plt_title = "smart-contract-data dataset 33767 samples"
# for ver in repo1_vers:
#   if ver not in sol_version_dict:
#     print(ver)
#   sol_version_dict[ver] += 1

### DApp
plt_title = "DApp dataset 21243 samples"
for ver in repo2_vers:
  if "<" in ver:
    continue
  if ver not in sol_version_dict:
    print(ver)
  sol_version_dict[ver] += 1

### BOTH
# plt_title = "both dataset 55010 samples"
```

```python
print(sol_version_dict)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# FIG INFO

versions = sol_version_dict

# PLOT 2
version_numbers = list(sol_version_dict.keys())
version_counts = list(sol_version_dict.values())

plt.figure(figsize=(20, 12))  # Adjust the figure size as needed
sns.barplot(x=version_numbers, y=version_counts, palette='viridis')
plt.xlabel('Version', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title(plt_title, fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels and adjust fontsize
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

# SORTED PLOT & REMOVE 0
versions_filtered = {k: v for k, v in versions.items() if v > 0}
sorted_versions = sorted(versions_filtered.items(), key=lambda x: x[1], reverse=True)
version_numbers = [item[0] for item in sorted_versions]
version_counts = [item[1] for item in sorted_versions]

plt.figure(figsize=(20, 6))
sns.barplot(x=version_numbers, y=version_counts, palette='viridis')
plt.xlabel('Version')
plt.ylabel('Count')
plt.title(plt_title)
plt.xticks(rotation=40, ha='right', fontsize=8)  # Rotate x-axis labels and adjust fontsize
plt.tight_layout()
plt.show()
```
